```python
# Código resultante para la implementación en Python

def primos_en_rango(limite_inferior, limite_superior):
    """
    Devuelve una lista con todos los números primos entre limite_inferior y limite_superior (inclusive).

    Parámetros:
    - limite_inferior (int): límite inferior del rango.
    - limite_superior (int): límite superior del rango.

    Retorno:
    - list[int]: lista de números primos ordenados de menor a mayor.
    """

    # Validar que los parámetros sean enteros
    if not isinstance(limite_inferior, int) or not isinstance(limite_superior, int):
        raise TypeError("Ambos límites deben ser números enteros.")

    # Ajustar el rango si los límites están invertidos
    if limite_inferior > limite_superior:
        limite_inferior, limite_superior = limite_superior, limite_inferior

    primos = []

    for num in range(max(2, limite_inferior), limite_superior + 1):
        es_primo = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                es_primo = False
                break
        if es_primo:
            primos.append(num)

    return primos


# Ejemplo de uso
print(primos_en_rango(10, 30))
```

```python
# Código automatizado de pruebas

def test_primos_en_rango():
    """
    Ejecuta casos de prueba para la función primos_en_rango().
    """
    test_cases = [
        ((10, 30), [11, 13, 17, 19, 23, 29]),
        ((2, 2), [2]),
        ((0, 1), []),
        ((30, 10), [11, 13, 17, 19, 23, 29]),
        ((-5, 5), [2, 3, 5]),
        ((14, 16), []),
        ((17, 17), [17]),
    ]

    excepciones = [
        (('a', 10), TypeError),
        ((10, 3.5), TypeError),
    ]

    # Pruebas de resultados
    for i, (inputs, expected) in enumerate(test_cases, 1):
        result = primos_en_rango(*inputs)
        assert result == expected, f"Fallo en caso {i}: esperado {expected}, obtenido {result}"
        print(f"Caso {i} pasado ✅")

    # Pruebas de excepciones
    for i, (inputs, exception) in enumerate(excepciones, 1):
        try:
            primos_en_rango(*inputs)
        except Exception as e:
            assert isinstance(e, exception), f"Fallo en excepción caso {i}: esperado {exception}, obtenido {type(e)}"
            print(f"Caso excepción {i} pasado ✅")
        else:
            assert False, f"Fallo en excepción caso {i}: no se lanzó {exception}"

    print("Todas las pruebas pasaron 🎉")


# Ejecutar pruebas
test_primos_en_rango()
```
